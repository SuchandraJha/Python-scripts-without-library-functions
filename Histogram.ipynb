```python
from PIL import Image
import matplotlib.pyplot as plt 

ImagePath="F:\Studies\JU studies\Histogram\Image6.jpg"
ImageText="txtImage6.txt"
```

```python
Im=Image.open(ImagePath)
plt.imshow(Im)
```

```python
Intensities=open(ImageText,"r")
PixelList=Intensities.readlines()
TotalPixels=len(PixelList)-3

# Finding rows and columns of the image (in string type)
strData=PixelList[2].split()
rows=strData[0]
columns=strData[1]

# Changing pixels from string type to integer type
for Loop in range(3,len(PixelList)):
    PixelList[Loop]=int(PixelList[Loop])

# Finding maximum number of pixels and storing it in 'max'
max=0
for Loop in range(3,len(PixelList)):
    if PixelList[Loop]>max:
        max=PixelList[Loop]
        
max=max+1
```

```python
# Counting no. of pixels for each intensity values and storing them in 'PlotList'
PlotList=[]
for Loop in range(0,max):
    PlotList.append(0)

for Loop in range(3,len(PixelList)):
    Index=PixelList[Loop]
    PlotList[Index]=PlotList[Index]+1
```

```python
# Storing each intensity in the list 'XValues'
XValues=[]
for Loop in range(0,max):
    XValues.append(0)
    
for Loop in range(0,max):
    XValues[Loop]=Loop
```

```python
plt.bar(XValues,PlotList)
```

```python
# Finding probability and storing them in 'NewPlotList' list
NewPlotList=[]
for Loop in range(0,max):
    NewPlotList.append(0)
    
for Loop in range(0,max):
    NewPlotList[Loop]=PlotList[Loop]/TotalPixels
```

```python
# Finding cumulative probabilities and replacing 'NewPlotList' with them
for Loop in range(1,max):
    NewPlotList[Loop]=NewPlotList[Loop]+NewPlotList[Loop-1]
```

```python
# Distributing the probabilities between 1-20 and replacing 'NewPlotList' with them
for Loop in range(1,max):
    NewPlotList[Loop]=NewPlotList[Loop]*max
```

```python
# Rounding to get number of pixels and replacing 'NewPlotList' with them
for Loop in range(1,max):
    NewPlotList[Loop]=int(NewPlotList[Loop])
```

```python
plt.bar(XValues,NewPlotList)
```

```python
# Replacing with revised number of pixels
for Loop in range(3,len(PixelList)):
    Index=PixelList[Loop]
    NewValue=NewPlotList[Index]
    PixelList[Loop]=NewValue
```

```python
# Writing new intensities in 'newfile.txt'
NewFile=open("newfile.txt","w+")

NewFile.write('P3\n# Created by GIMP version 2.10.18 PNM plug-in\n'+rows+' '+columns+'\n')

for Loop in range(3,len(PixelList)):
    val=str(PixelList[Loop])
    NewFile.write(val)
    NewFile.write('\n')

NewFile.write('\n')
```

```python
# Histogram of new image
ImageText="newfile.txt"

Intensities=open(ImageText,"r")
PixelList=Intensities.readlines()

# Changing pixels from string type to integer type
for Loop in range(3,len(PixelList)):
    PixelList[Loop]=int(float(PixelList[Loop]))

# Finding maximum number of pixels and storing it in 'max'
max=0
for Loop in range(3,len(PixelList)):
    if PixelList[Loop]>max:
        max=PixelList[Loop]
        
max=max+1

# Counting no. of pixels for each intensity values and storing them in 'PlotList'
PlotList=[]
for Loop in range(0,max):
    PlotList.append(0)

for Loop in range(3,len(PixelList)):
    Index=PixelList[Loop]
    PlotList[Index]=PlotList[Index]+1
    
# Storing each intensity in the list 'XValues'
XValues=[]
for Loop in range(0,max):
    XValues.append(0)
    
for Loop in range(0,max):
    XValues[Loop]=Loop
    
plt.bar(XValues,PlotList)
```
